# Week 3 Answer - Regression

```python
import random

random.seed(0)

# Generate a univariate dataset with 100 data points
X=[]
r=[]
N=100
for i in range(N):
    X.append(random.random())
    r.append(random.random())

print('dataset %s' %list(zip(X,r)))
```

```python
# calculate w_1 and w_0 for the dataset
N=len(X)
mean_X = sum(X)/N
mean_r = sum(r)/N

w1_numerator = 0
for index, x in enumerate(X):
    w1_numerator += x * r[index]

w1_numerator -= (mean_X*mean_r*N)

w1_denominator = 0
for index, x in enumerate(X):
    w1_denominator += pow(x,2)

w1_denominator = N*pow(mean_X,2)

w1 = w1_numerator/w1_denominator

w0 = mean_r - w1*mean_X

print('w1: %s' %w1)
print('w0: %s' %w0)
```

```python
# Report the error with the calculated w_1 and w_0

def g(x, w1, w0):
    return w1*x + w0

# report Root Mean Squared Error (RMSE) as output
sum_squared_error=0
for index, x in enumerate(X):
    error = g(x, w1, w0) - r[index]
    squared_error = pow(error, 2)
    sum_squared_error += squared_error

mean_squared_error = sum_squared_error/N
RMSE = pow(mean_squared_error, 0.5)
print('RMSE: %s' %RMSE)
```

```python
# Plot the data points along with the fitted line
import matplotlib.pyplot as plt
plt.plot(X, r, 'ro')
plt.plot(X, [g(x, w1, w0) for x in X] )
plt.show()
```
